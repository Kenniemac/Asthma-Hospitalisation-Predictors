```python

from sklearn.impute import SimpleImputer

header_names = ['City','Ozone','Tox Releases', 'Asthma', 'Education', 'Poverty',
       'Children < 10 (%)', 'Pop 11-64 years (%)', 'Elderly > 65 (%)',
       'Hispanic (%)', 'White (%)', 'African American (%)',
       'Native American (%)', 'Asian American (%)', 'Other (%)',
       '2019 Smoking Products']

#use the simple imputer function to replace missing value
imputer = SimpleImputer(strategy='most_frequent')
imputer.fit(df)
new_data = imputer.transform(df)

df= pd.DataFrame(data=new_data, columns=header_names)

df.head()
#check for missing values again or still present
empty_data = df[df.isna().any(axis=1)]
print('\nThese are the missing data \n', empty_data) #you may note there is no missing data
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from pandas.plotting import scatter_matrix
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split,cross_val_score, KFold
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
X = df.drop(['Asthma'], axis=1)
y= df['Asthma']

X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=1)

#check the dimensions of the train and test data
print('\n The total number of training dataset', X_train.shape)
print(('\n The total number of test dataset', X_test.shape))
```

```python
#performing random forest regression
rf_model = RandomForestRegressor(n_estimators=25, random_state=2)

#fit bc to the training dataset
rf_model.fit(X_train, y_train)

#predict the labels f the training set
y_pred_train = rf_model.predict(X_train)

#predict test set labels
y_pred = rf_model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

# the r2 for the train dataset
r2_train = r2_score(y_test, y_pred)

print('Root mean squared error: {:2f}'.format(rmse))
print('Mean squared error:{:2f}'.format(mse))
print('Mean absolute error:{:2f}'.format(mae))
print('R2 score:{:.2f}'.format(r2))
print('R2 score Train data: {:.2f}'.format(r2_train))
                   
```

```python
#visualise feature importance 
#Create a pd.Series of importance
importances = pd.Series(data=rf_model.feature_importances_, index= X_train.columns)

# Sort based on prediction importances
importances_sorted = importances.sort_values()

#Draw a horizontal barplot of importance sorted
importances_sorted.plot(kind='barh', color='blue')
plt.title('Features Importances')
plt.show()
```

```python
from sklearn.ensemble import AdaBoostRegressor

dt_model = DecisionTreeRegressor(max_depth=2, random_state=1)

#instantiate ada
ada=AdaBoostRegressor(base_estimator=dt_model, n_estimators=180, random_state=1)


#fit ada to the training set
ada.fit(X_train, y_train, y_train.values.ravel())

#predict the labels of the training
y_pred_train = ada.predict(X_train)

#predict model
y_pred= ada.predict(X_test)

#evaluate the model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

# the r2 for the train dataset
r2_train = r2_score(y_test, y_pred)

print('Root mean squared error: {:2f}'.format(rmse))
print('Mean squared error:{:2f}'.format(mse))
print('Mean absolute error:{:2f}'.format(mae))
print('R2 score:{:.2f}'.format(r2))
print('R2 score Train data: {:.2f}'.format(r2_train))
```
